```python
# read data from csv file
df = pd.read_csv('main_task.csv')

# change columns names
df.columns = ['id', 'city_name', 'cuisine', 'rank', 'rating', 'price', 'reviews count', 'reviews', 'url_ta', 'id_ta']

# convert id to number
df['id'] = df['id'].str[3:].astype(int)

# city
df.insert(1, 'city_name_length', df['city_name'].str.len())
df.insert(2, 'city_id', 0)
df.insert(3, 'country_id', 0)
df.insert(4, 'is_in_capital', 0)
df.insert(5, 'far_from_capital', 0)
df.insert(6, 'city_population', 0)

df_cities = df['city_name'].value_counts().reset_index()
df_cities.columns = ['name', 'count']

cities = {}

count = df_cities['name'].size
for i in range(0, count):
    city = df_cities.loc[i]
    cities[city['name']] = {
        'id': int(i),
        'count': city['count']
    }
    
countries_ids = {
    'UK': 0,
    'France': 1,
    'Spain': 2,
    'Germany': 3,
    'Italy': 4,
    'Czech Republic': 5,
    'Portugal': 6,
    'Austria': 7,
    'Netherlands': 8,
    'Belgium': 9,
    'Sweden': 10,
    'Hungary': 11,
    'Poland': 12,
    'Ireland': 13,
    'Denmark': 14,
    'Greece': 15,
    'Switzerland': 16,
    'Norway': 17,
    'Finland': 18,
    'Slovakia': 19,
    'Luxembourg': 20,
    'Slovenia': 21,
}

cities['London'].update({
    'country_id': countries_ids['UK'],
    'is_capital': 1,
    'far_from_capital': 0,
    'population': 8.9,
})
cities['Paris'].update({
    'country_id': countries_ids['France'],
    'is_capital': 1,
    'far_from_capital': 0,
    'population': 2.1,
})
cities['Madrid'].update({
    'country_id': countries_ids['Spain'],
    'is_capital': 1,
    'far_from_capital': 0,
    'population': 6.6,
})
cities['Barcelona'].update({
    'country_id': countries_ids['Spain'],
    'is_capital': 0,
    'far_from_capital': 504,
    'population': 5.6,
})
cities['Berlin'].update({
    'country_id': countries_ids['Germany'],
    'is_capital': 1,
    'far_from_capital': 0,
    'population': 3.8,
})
cities['Milan'].update({
    'country_id': countries_ids['Italy'],
    'is_capital': 0,
    'far_from_capital': 477,
    'population': 1.3,
})
cities['Rome'].update({
    'country_id': countries_ids['Italy'],
    'is_capital': 1,
    'far_from_capital': 0,
    'population': 2.9,
})
cities['Prague'].update({
    'country_id': countries_ids['Czech Republic'],
    'is_capital': 1,
    'far_from_capital': 0,
    'population': 10.7,
})
cities['Lisbon'].update({
    'country_id': countries_ids['Portugal'],
    'is_capital': 1,
    'far_from_capital': 0,
    'population': 0.5,
})
cities['Vienna'].update({
    'country_id': countries_ids['Austria'],
    'is_capital': 1,
    'far_from_capital': 0,
    'population': 1.9,
})
cities['Amsterdam'].update({
    'country_id': countries_ids['Netherlands'],
    'is_capital': 1,
    'far_from_capital': 0,
    'population': 0.8,
})
cities['Brussels'].update({
    'country_id': countries_ids['Belgium'],
    'is_capital': 1,
    'far_from_capital': 0,
    'population': 2,
})
cities['Hamburg'].update({
    'country_id': countries_ids['Germany'],
    'is_capital': 0,
    'far_from_capital': 255,
    'population': 1.8,
})
cities['Munich'].update({
    'country_id': countries_ids['Germany'],
    'is_capital': 0,
    'far_from_capital': 255,
    'population': 1.5,
})
cities['Lyon'].update({
    'country_id': countries_ids['France'],
    'is_capital': 0,
    'far_from_capital': 393,
    'population': 1.7,
})
cities['Stockholm'].update({
    'country_id': countries_ids['Sweden'],
    'is_capital': 1,
    'far_from_capital': 0,
    'population': 1.6,
})
cities['Budapest'].update({
    'country_id': countries_ids['Hungary'],
    'is_capital': 1,
    'far_from_capital': 0,
    'population': 1.8,
})
cities['Warsaw'].update({
    'country_id': countries_ids['Poland'],
    'is_capital': 1,
    'far_from_capital': 0,
    'population': 1.8,
})
cities['Dublin'].update({
    'country_id': countries_ids['Ireland'],
    'is_capital': 1,
    'far_from_capital': 0,
    'population': 1.2,
})
cities['Copenhagen'].update({
    'country_id': countries_ids['Denmark'],
    'is_capital': 1,
    'far_from_capital': 0,
    'population': 1.3,
})
cities['Athens'].update({
    'country_id': countries_ids['Greece'],
    'is_capital': 1,
    'far_from_capital': 0,
    'population': 3.2,
})
cities['Edinburgh'].update({
    'country_id': countries_ids['UK'],
    'is_capital': 0,
    'far_from_capital': 534,
    'population': 0.5,
})
cities['Zurich'].update({
    'country_id': countries_ids['Switzerland'],
    'is_capital': 0,
    'far_from_capital': 95,
    'population': 1.4,
})
cities['Oporto'].update({
    'country_id': countries_ids['Portugal'],
    'is_capital': 0,
    'far_from_capital': 300,
    'population': 1.3,
})
cities['Geneva'].update({
    'country_id': countries_ids['Switzerland'],
    'is_capital': 0,
    'far_from_capital': 130,
    'population': 1.3,
})
cities['Krakow'].update({
    'country_id': countries_ids['Poland'],
    'is_capital': 0,
    'far_from_capital': 252,
    'population': 0.8,
})
cities['Oslo'].update({
    'country_id': countries_ids['Norway'],
    'is_capital': 1,
    'far_from_capital': 0,
    'population': 0.7,
})
cities['Helsinki'].update({
    'country_id': countries_ids['Finland'],
    'is_capital': 1,
    'far_from_capital': 0,
    'population': 1.3,
})
cities['Bratislava'].update({
    'country_id': countries_ids['Slovakia'],
    'is_capital': 1,
    'far_from_capital': 0,
    'population': 0.4,
})
cities['Luxembourg'].update({
    'country_id': countries_ids['Luxembourg'],
    'is_capital': 1,
    'far_from_capital': 0,
    'population': 0.6,
})
cities['Ljubljana'].update({
    'country_id': countries_ids['Slovenia'],
    'is_capital': 1,
    'far_from_capital': 0,
    'population': 0.3,
})

for city_name, city_data in cities.items():
    df.loc[df['city_name'] == city_name, 'city_id'] = int(city_data['id'])
    df.loc[df['city_name'] == city_name, 'country_id'] = int(city_data['country_id'])
    df.loc[df['city_name'] == city_name, 'is_in_capital'] = int(city_data['is_capital'])
    df.loc[df['city_name'] == city_name, 'far_from_capital'] = int(city_data['far_from_capital'])
    df.loc[df['city_name'] == city_name, 'city_population'] = int(city_data['population'])
    
df.drop('city_name', axis='columns', inplace=True)

# cuisine
df['cuisine'] = df['cuisine'].fillna("['Vegetarian Friendly', 'European']")
df['cuisine'] = df['cuisine'].str.replace("'", '').str.replace("\[", '').str.replace("\]", '').str.split(', ');

df.insert(8, 'cuisines_count', None)
df['cuisines_count'] = df['cuisine'].apply(lambda c: len(c))

cuisines = set(df['cuisine'].explode().tolist())
for cuisine_name in cuisines:
    column_name = 'cuisine_' + cuisine_name.lower().replace(' ', '_')
    df[column_name] = df['cuisine'].apply(lambda cuisines: int(cuisine_name in cuisines))
    
df.drop('cuisine', axis='columns', inplace=True)

# price
def price_zone(price):
        if price == '$':
            return 0
        elif price == '$$ - $$$':
            return 1
        elif price == '$$$$':
            return 2
            
    
df['price'].fillna('$$ - $$$', inplace=True)
df['price'] = df['price'].apply(price_zone).astype(int)

# reviews
def split_reviews_dates(string):
    split_position = string.find('], [')
    reviews_end_position = split_position
    dates_start_position = split_position + 4
    reviews_string = string[0:reviews_end_position].replace('[', '')
    dates_string = string[dates_start_position:].replace(']', '')
    reviews = list(map(lambda review: review[1:-1], reviews_string.split(', ')))
    dates = list(map(lambda date: date[1:-1], dates_string.split(', ')))
    reviews = list(filter(lambda item: len(item) > 0, reviews))
    dates = list(filter(lambda item: len(item) > 0, dates))
    dates = list(map(lambda item: datetime.strptime(item, '%m/%d/%Y'), dates))
    return reviews, dates


def get_reviews_list(string):
    return split_reviews_dates(string)[0]


def get_dates_list(string):
    return split_reviews_dates(string)[1]


def get_last_date(string):
    dates = get_dates_list(string)
    if len(dates) > 0:
        return max(dates)
    else:
        return datetime.strptime('01/01/1971', '%m/%d/%Y')


def get_time_between(string):
    dates = get_dates_list(string)
    if len(dates) > 1:
        diffs = []
        for i in range(1, len(dates)):
            diffs.append(abs((dates[i] - dates[i-1]).days))
        return  stat.mean(diffs)
    else:
        return 0


def get_time_from_last_review(string):
    last_date = get_last_date(string)
    now = datetime.now()
    from_last = now - last_date
    return from_last.days
    
    
df['reviews'] = df['reviews'].str.replace('\"', '').str.replace('\"', '');

df.insert(12, 'reviews_count', None)
df['reviews_count'] = df['reviews'].map(lambda reviews: len(get_reviews_list(reviews)))

df.insert(13, 'time_between_reviews', None)
df.insert(14, 'time_from_last_review', None)
df['time_between_reviews'] = df['reviews'].map(lambda reviews: get_time_between(reviews))
df['time_from_last_review'] = df['reviews'].map(get_time_from_last_review)

df.drop('reviews', axis='columns', inplace=True)

# url_ta
# df.drop('url_ta', axis = 'columns', inplace = True)

# id_ta
df.drop('id_ta', axis = 'columns', inplace = True)

display(df.loc[0,'url_ta'])  
```
